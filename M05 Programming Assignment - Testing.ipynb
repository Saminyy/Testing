```python
#test_sum.py
def test_sum():
    assert sum([1, 2, 3]) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum()
    print("Everything passed")
```

The above is a test that checks the output of the function with a known output. If it does not equal the known output an error is raised. If it is equal to the known output it prints a confirmation message.

```python
#test_sum_2.py
def test_sum():
    assert sum([1, 2, 3]) == 6, "Should be 6"

def test_sum_tuple():
    assert sum((1, 2, 2)) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum()
    test_sum_tuple()
    print("Everything passed")
```

The above test checks against a tuple in the same way as the previous test. This raises an error because it equals 5 not 6.

```python
#test_sum_unittest.py
import unittest

class TestSum(unittest.TestCase):

    def test_sum(self):
        self.assertEqual(sum([1, 2, 3]), 6, "Should be 6")

    def test_sum_tuple(self):
        self.assertEqual(sum((1, 2, 2)), 6, "Should be 6")

if __name__ == '__main__':
    unittest.main()
```

The test shows it successfully ran both tests with the results printed at the top. The '.' means the first test passed and the 'F' means the second test failed. In the event of a test failing, the test results will also display details on where the error was and what result it was expecting to see, which it what shows in the results above indicating line 10 is where the error of not receiving the expected output happens.

```python
import unittest

from my_sum import sum
from fractions import Fraction

class TestSum(unittest.TestCase):
    def test_list_int(self):
        """
        Test that it can sum a list of integers
        """
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)
    def test_list_fraction(self):
        """
        Test that it can sum a list of fractions
        """
        data = [Fraction(1, 4), Fraction(1, 4), Fraction(2, 5)]
        result = sum(data)
        self.assertEqual(result, 1)
    

if __name__ == '__main__':
    unittest.main()
```

The test results will show either an '.' or an 'F'. The '.' means the test passed, the 'F' means the test failed. In the event of a test failing, the test results will also display details on where the error was and what result it was expecting to see.
